```python
from urllib.request import urlopen


def download(url):
# Doanloads raw data and module from GitHub
    filename = url.split('/')[-1]
    f = urlopen(url)
    data = f.read()
    f.close()
    with open(filename, 'wb') as myfile:
        myfile.write(data)

# Get repository
download('https://raw.githubusercontent.com/khyatiparekh/Data515_FinalProject/master/SearchingForSuccess/Submodule/EarningReport.py')
download('https://raw.githubusercontent.com/khyatiparekh/Data515_FinalProject/master/SearchingForSuccess/Submodule/PredictionStockPrice.py')
download('https://raw.githubusercontent.com/khyatiparekh/Data515_FinalProject/master/SearchingForSuccess/Submodule/HistoricalDataPlot.py')
download('https://raw.githubusercontent.com/khyatiparekh/Data515_FinalProject/master/SearchingForSuccess/Submodule/Amazon.csv')

# Import Modules
import EarningReport
import PredictionStockPrice
import HistoricalDataPlot

# Example Output
EarningReport.get_current_quarter_dates()
EarningReport.show_last_day_quarter()
EarningReport.get_earnings_data("Amazon")
print("Predicted probability the stock price will increase after releasing next quarterâ€™s earnings report: ")
print(PredictionStockPrice.prediction_stock_price("Amazon"))
HistoricalDataPlot.historical_data_plot("Amazon")
```
